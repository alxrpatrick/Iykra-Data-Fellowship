# Exploratory Data Analysis

### Importing libraries

```python
import pandas as pd
import numpy as np
```

```python
# set to view all columns
pd.set_option('max_columns', None)
```

### Importing Dataset

```python
epl_data = pd.read_csv('epl_1819.csv')
```

```python
epl_data
```

## Data Preprocessing

### Getting tables with null value

```python
epl_data.isna().sum()
```

### Getting datatypes of all columns

```python
epl_data.dtypes
```

### Removing commas from numerical values and casting datatype to integer

```python
epl_data['attack_passes'] = epl_data['attack_passes'].str.replace(',', '').astype(int)
epl_data['attack_passes_long'] = epl_data['attack_passes_long'].str.replace(',', '').astype(int)
epl_data['attack_passes_back'] = epl_data['attack_passes_back'].str.replace(',', '').astype(int)
epl_data['defence_clearances'] = epl_data['defence_clearances'].str.replace(',', '').astype(int)
```

```python
epl_data
```

```python
epl_data.dtypes
```

## Getting Data Description

```python
epl_data['general_points'].describe()
```

```python
# getting upper and lower bound values
Q3 = np.percentile(epl_data['general_points'],75)
Q1 = np.percentile(epl_data['general_points'],25)
IQR = Q3 - Q1
UB = Q3 + 1.5 * IQR
LB = Q1 - 1.5 * IQR
print('UB : %f\nLB : %f'%(UB, LB))
```

### Box plot for distribution of the data

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (8,10))
sns.boxplot(data=epl_data['general_points'])
```

## Getting Team with Best Attack Stats

```python
# creating seperate dataframe
attack_stat = epl_data.loc[:, 'attack_passes':'attack_goals_outsidebox']
```

```python
attack_stat
```

```python
# append columns
attack_stat = pd.concat([epl_data['Team'], attack_stat, epl_data['attack_scored'], epl_data['attack_goals_counter'], epl_data['attack_goals_freekick']], axis=1)
```

### Getting mean value for each team

```python
attack_stat['attack_stat_avg'] = attack_stat.mean(axis=1)
```

```python
attack_stat
```

### Getting team with largest average

```python
attack_stat[attack_stat['attack_stat_avg'] == attack_stat['attack_stat_avg'].max()]
```

### Plotting the data

```python
fig, ax = plt.subplots(figsize=(5,8))
sns.barplot(x="attack_stat_avg", y="Team", ax = ax, data=attack_stat)
```

## Getting team with best defence stats

```python
# creating seperate dataframe
defence_stat = epl_data.loc[:, 'defence_saves':'defence_clearances_headed']
```

```python
# append columns
defence_stat = pd.concat([epl_data['Team'], defence_stat], axis=1)
```

```python
# getting average for each team
defence_stat['defence_stat_avg'] = defence_stat.mean(axis=1)
```

```python
defence_stat
```

### Getting team with largest average

```python
defence_stat[defence_stat['defence_stat_avg'] == defence_stat['defence_stat_avg'].max()]
```

### Plotting the data

```python
fig, ax = plt.subplots(figsize=(5,8))
sns.barplot(x="defence_stat_avg", y="Team", ax = ax, data=defence_stat)
```

## Getting team with best financial aspect

```python
# creating separate dataframe
finance_stat = epl_data[['Team','finance _tv_revenue', 'finance _team_market', 'finance _market_average']]
```

```python
# getting sum
finance_stat['net_worth'] = finance_stat.sum(axis=1)
```

```python
finance_stat
```

### Getting team with most net worth

```python
finance_stat[finance_stat['net_worth'] == finance_stat['net_worth'].max()]
```

### Plotting the data

```python
fig, ax = plt.subplots(figsize=(5,8))
sns.barplot(x="net_worth", y="Team", ax = ax, data=finance_stat)
```
