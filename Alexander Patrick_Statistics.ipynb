### Importing libraries and dataset

```python
import pandas as pd
```

```python
stats = pd.read_csv('Seasons_Stats.csv',header=0)
```

```python
stats
```

## Data Preprocessing

Here, we're only going to use data from year 2017

```python
stats_2017 = stats[stats['Year'] == 2017]
```

```python
stats_2017
```

### Remove Duplicates
There are some players who make team transfers in the NBA transfer market so that there is duplication of player data. Therefore, we will remove duplicated players and keep the most recent data

```python
stats_2017 = stats_2017.drop_duplicates('Player', keep='last')
stats_2017
```

### Deleting null values

There are also some null values in the data which will be eliminated

```python
stats_2017.isnull().sum()
```

From the description above, 2 columns (blanl and blank2) are  having no value at all which makes them redundant to be on the dataset. Hence they will be removed

```python
stats_2017 = stats_2017.drop(['blanl', 'blank2'], axis=1)
stats_2017
```

## 1. The youngest and oldest player in the NBA in 2017 for each team 
Data of the youngest/oldest player for each team is obtained by grouping the data by team attribute, and using transform to get the minimum/maximum data of each team

### Youngest

```python
# find result of each team by group by function
youngest = stats_2017[stats_2017.groupby('Tm')['Age'].transform(min) == stats_2017['Age']]
youngest[['Tm', 'Player', 'Age']].reset_index()
```

### Oldest

```python
oldest = stats_2017[stats_2017.groupby('Tm')['Age'].transform(max) == stats_2017['Age']]
oldest[['Tm', 'Player', 'Age']].reset_index()
```

## 2. Player that has the most minutes played (MP) in each position 

Data for player with most minutes played in each position will be gathered by grouping the dataset by position column and using transform function to get the maximum minutes from MP column

```python
max_minute = stats_2017[stats_2017.groupby('Pos')['MP'].transform(max) == stats_2017['MP']]
max_minute[['Pos', 'Player', 'MP']].reset_index()
```

## 3. Team that has the highest average total rebound percentage (TRB%), assist percentage (AST%), steal percentage (STL%), and block percentage (BLK%)

### Creating Seperate Table

```python
highest_avg = stats_2017[['Tm', 'TRB%', 'AST%', 'STL%', 'BLK%']]
```

Data will be gathered by getting average total of each team for each attribute

```python
avg_team = highest_avg.groupby('Tm').mean().reset_index()
avg_team
```

### The maximum value for each attribute will be obtained by getting the entity having the maximum value of the attribute

### Rebound (TRB%)

```python
mean_trb = avg_team[['Tm', 'TRB%']]
mean_trb_max = mean_trb[mean_trb['TRB%'] == mean_trb['TRB%'].max()]
mean_trb_max
```

### Assist (AST%)

```python
mean_ast = avg_team[['Tm', 'AST%']]
mean_ast_max = mean_ast[mean_ast['AST%'] == mean_ast['AST%'].max()]
mean_ast_max
```

### Steal (STL%)

```python
mean_stl = avg_team[['Tm', 'STL%']]
mean_stl_max = mean_stl[mean_stl['STL%'] == mean_stl['STL%'].max()]
mean_stl_max
```

### Block (BLK%)

```python
mean_blk = avg_team[['Tm', 'BLK%']]
mean_blk_max = mean_blk[mean_blk['BLK%'] == mean_blk['BLK%'].max()]
mean_blk_max
```

## 4. The best player in my opinion based on his record stats

```python
# filling null values with 0
stats_2017 = stats_2017.fillna(0)
```

There are some parameters used to determine who is the best player based on his record stats. 
Some stats per game metrics are collected and the average of all metrics will be calculated. Player with highest number of average will be nominated as the best player.
Here are some stats used as parameters:<br>
    - GS  : Game Starred <br>
    - PER : Player Efficiency Rating <br>
    - WS  : Win Shares<br>
    - FG  : Field Goals<br>
    - 3p  : 3 points field goal<br>
    - 2p  : 2 points field goal<br>
    - ORB : Offensive Rebounds<br>
    - DRB : Defensive Rebounds<br>
    - TRB : Total Rebounds<br>
    - AST : Assists<br>
    - STL : Steals<br>
    - BLK : Blocks<br>
    - PTS : Points<br>

```python
# create seperate table
best_stat = stats_2017[['Player', 'Tm', 'GS', 'PER', 'WS', 'FG', '3P', '2P',
                        'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'PTS']].reset_index(drop=True)
best_stat
```

```python
# creating record average column
best_stat['Record_Avg'] = best_stat.mean(axis=1)
```

```python
best_stat
```

The max value shows the nominated best player in 2017

```python
# getting maximum value
best_stat[best_stat['Record_Avg'] == best_stat['Record_Avg'].max()]
```

## 5. Team with the best average stat record of their players

The best team data will be obtained by the same parameter in number 4. Data will be grouped by the team and the mean of the record average will be calculated for each team. 

```python
best_team = best_stat.groupby('Tm').mean().reset_index()
```

```python
best_team
```

The team with the maximum value of the record average will be nominated the best team

```python
best_team[best_team['Record_Avg'] == best_team['Record_Avg'].max()]
```
